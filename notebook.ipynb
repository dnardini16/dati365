#### CSV

Un **CSV** (Comma-Separated Values) è un formato di file semplice utilizzato per memorizzare dati tabulari, come una tabella o un foglio di calcolo. Ogni riga nel file rappresenta un record, e ogni record è composto da uno o più campi, separati da un carattere delimitatore (tipicamente una virgola, da cui deriva il nome "Comma-Separated Values"). I file CSV sono ampiamente utilizzati perché sono leggeri, facili da leggere e possono essere aperti da una vasta gamma di software, tra cui fogli di calcolo come Microsoft Excel, Google Sheets e librerie di programmazione come Pandas in Python.


Esempio di CSV
```
ID_Prodotto,Nome,Prezzo,Quantità
1,Prodotto A,10.5,100
2,Prodotto B,7.8,200
3,Prodotto C,15.0,150
```

#### ENCODING

L'**encoding** è il processo di conversione di caratteri in un file di testo in un formato specifico che può essere letto e interpretato dai computer. Questo formato di codifica traduce ogni carattere in una sequenza di byte. Diversi standard di encoding esistono per rappresentare caratteri da diverse lingue e simboli speciali, il che è particolarmente importante in contesti globali.

Alcuni degli encoding più comuni includono:

- UTF-8: È uno dei formati di encoding più usati e compatibili. Supporta un'ampia gamma di caratteri da molte lingue, mantenendo la compatibilità con il set di caratteri ASCII (che è limitato ai caratteri latini di base).
- ISO-8859-1 (Latin-1): Supporta i caratteri latini di base utilizzati in molte lingue europee occidentali.
- ASCII: Un encoding molto semplice che supporta solo i caratteri inglesi di base, senza supporto per caratteri accentati o simboli speciali non inglesi.

# ANALISI ESPLORATIVA

L'**Analisi Esplorativa dei Dati** (EDA) è un processo fondamentale nel data science e nella statistica che coinvolge l'analisi iniziale dei dati per riassumerne le principali caratteristiche, spesso con l'ausilio di metodi visivi. L'obiettivo dell'EDA è comprendere la struttura dei dati, individuare modelli, anomalie, relazioni, e formare ipotesi che possano guidare ulteriori analisi.

Obiettivi dell'EDA:

- Comprendere la Distribuzione dei Dati: Identificare come i dati sono distribuiti, ad esempio se seguono una distribuzione normale o se ci sono valori anomali.

- Identificare le Relazioni tra le Variabili: Esplorare come le diverse variabili del dataset interagiscono tra loro, cercando correlazioni e associazioni.

- Individuare Anomalie e Outlier: Rilevare valori anomali o errori nei dati che potrebbero distorcere l'analisi.

- Preparazione per Modelli Predittivi: Pulire e trasformare i dati per renderli adatti all'analisi statistica o alla modellazione predittiva.

# ANALISI DELLE VENDITE

````python
# merge dei dataset transazioni e clienti

transazioni.rename(columns={'id cliente': 'ID Cliente'}, inplace=True)
clienti.rename(columns={'ID cliente': 'ID Cliente'}, inplace=True)

dati_completi = pd.merge(transazioni, clienti, on='ID Cliente', how='left')
````

````python
dati_completi.head()
````

# CALCOLO DEL FATTURATO

````python
## controllo per evitare "sorprese"
dati_completi["prezzo d'acquisto"] = dati_completi["prezzo d'acquisto"].apply(lambda row: row.replace(",", "."))
````

````python
## conversione in float per le operazioni matematiche
dati_completi["prezzo d'acquisto"] = pd.to_numeric(dati_completi["prezzo d'acquisto"])
dati_completi["quantita"] = pd.to_numeric(dati_completi["quantita"])

fatturato_totale = dati_completi["prezzo d'acquisto"] * dati_completi['quantita']
dati_completi['Fatturato'] = fatturato_totale
````

````python
dati_completi.head()
````

### ANALISI DEI PRODOTTI

````python
# merge dei dataset dati_completi e anagrafica

dati_completi.rename(columns={'id prodotto': 'ID prodotto'}, inplace=True)
dati_completi = pd.merge(dati_completi, anagrafica, on='ID prodotto', how='left')
````

````python
dati_completi.head()
````

````python
## prodotti più acquistati
fatturato_per_prodotto = dati_completi.groupby('Nome prodotto')['Fatturato'].sum()
fatturato_per_prodotto.sort_values(ascending=False).head()
````

````python
## prodotti meno acquistati
fatturato_per_prodotto.sort_values(ascending=False).tail()
````

````python
## categorie più acquistate
fatturato_per_prodotto = dati_completi.groupby('Categoria prodotto')['Fatturato'].sum()
fatturato_per_prodotto.sort_values(ascending=False).head()
````

````python
## categorie meno acquistate
fatturato_per_prodotto = dati_completi.groupby('Categoria prodotto')['Fatturato'].sum()
fatturato_per_prodotto.sort_values(ascending=False).tail()
````

````python
# Visualizzazione dei risultati
plt.figure(figsize=(30, 10))
sns.barplot(x=fatturato_per_prodotto.index, y=fatturato_per_prodotto.values)
plt.xlabel('Categoria Prodotto')
plt.ylabel('Fatturato')
plt.show()
````

### ANALISI RECENSIONI

````python
## merge con recensioni

dati_completi = pd.merge(dati_completi, recensioni[["ID prodotto", "Rating recensione"]], on='ID prodotto', how='left')
dati_completi.head()
````

````python
## prodotti con migliori recensioni
rating_per_prodotto = dati_completi.groupby('Nome prodotto')['Rating recensione'].mean()
rating_per_prodotto.sort_values(ascending=False).head()
````

````python
## prodotti con peggiori recensioni
rating_per_prodotto.sort_values(ascending=False).tail()
````
